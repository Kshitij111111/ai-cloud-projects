```python
import pandas as pd

# Load the uploaded CSV
df = pd.read_csv('Sales.csv')

# Show original data
print("üîπ Original Data:")
print(df)

# Check missing values
print("\nüîç Missing Values:")
print(df.isnull().sum())

# Fill missing values
df['Total_Sales'] = df['Total_Sales'].fillna(df['Units_Sold'] * df['Unit_Price'])
df['Unit_Price'] = df['Unit_Price'].fillna(df['Total_Sales'] / df['Units_Sold'])
df['Units_Sold'] = df['Units_Sold'].fillna(df['Total_Sales'] / df['Unit_Price'])

# Cleaned data
print("\n‚úÖ Cleaned Data:")
print(df)

# Save cleaned CSV
df.to_csv('sales_cleaned.csv', index=False)
```
