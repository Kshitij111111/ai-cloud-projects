```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Clean visuals
sns.set(style='whitegrid')

```

```python
from google.colab import files
uploaded = files.upload()  # Upload your Sales.csv file

df = pd.read_csv('Sales.csv.txt')
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.dropna()  # or use fillna if needed
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print(top_products)
top_products.plot(kind='bar', title='Top 5 Products by Sales')
plt.ylabel('Total Sales')
plt.show()
```

```python
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total_Sales'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend', figsize=(12,6))
plt.ylabel('Total Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_table = pd.pivot_table(df, values='Total_Sales', index='Product', columns='Month', aggfunc='sum')
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='Blues')
plt.title('Sales Heatmap by Product and Month')
plt.show()
```
