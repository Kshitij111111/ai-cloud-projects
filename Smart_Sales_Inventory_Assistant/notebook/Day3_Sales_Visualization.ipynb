```python
df = df.dropna()  # or use fillna if needed
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print(top_products)
top_products.plot(kind='bar', title='Top 5 Products by Sales')
plt.ylabel('Total Sales')
plt.show()
```

```python
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Total_Sales'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend', figsize=(12,6))
plt.ylabel('Total Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_table = pd.pivot_table(df, values='Total_Sales', index='Product', columns='Month', aggfunc='sum')
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='Blues')
plt.title('Sales Heatmap by Product and Month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Sales.csv.txt')  # Adjust path if needed

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
```

```python
# Total daily sales
daily_sales = df.groupby('Date')['Total_Sales'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Total_Sales', data=daily_sales)
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Total_Sales'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='Month', y='Total_Sales', data=monthly_sales)
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
product_sales = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
product_sales.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df['Day'] = df['Date'].dt.day
df['MonthName'] = df['Date'].dt.strftime('%B')

heatmap_data = df.pivot_table(index='Day', columns='MonthName', values='Total_Sales', aggfunc='sum')

plt.figure(figsize=(12,8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=False)
plt.title("Sales Heatmap by Day and Month")
plt.show()
```
